## Importing the necessary libraries

```python
import glob
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

## Stage 1

```python
DF_COMPANIES = (pd.concat([pd.read_csv(f, index_col='Date')['Close'].rename(f.split('/')[-1][6:-4])
                for f in sorted(glob.glob('stock/*.csv'))], axis=1))
```

```python
DF_COMPANIES
```

## Stage 2

```python
DF_CORR = DF_COMPANIES.corr()
```

```python
DF_CORR
```

## Stage 3

```python
plt.figure(figsize=(6, 5))
sns.heatmap(DF_CORR, cmap='coolwarm')
plt.title('Correlation matrix')
plt.show()
```

**Based on the correlation matrix, it can be concluded that positive correlation predominates between the stocks of the companies.**

## Stage 4

```python
googl_corr = DF_CORR.loc['GOOGL']
LST_COMP_CORR = [
    googl_corr.drop('GOOGL').idxmax(),
    googl_corr.idxmin(),
    googl_corr.abs().idxmin()
]
```

```python
LST_COMP_CORR
```

## Stage 5

```python
plt.scatter(x=DF_COMPANIES[LST_COMP_CORR[0]], y=DF_COMPANIES['GOOGL'], color='green')
plt.xlabel(LST_COMP_CORR[0])
plt.ylabel('GOOGL')
plt.title('Prices')
plt.grid()
plt.show()
```

**The graph shows the relationship between ADBE (Adobe) and GOOGL (Google) stock prices during the specified period. It confirms a positive correlation between them: when one stock's price rises, the other's typically increases as well.**

```python
plt.scatter(x=DF_COMPANIES[LST_COMP_CORR[1]], y=DF_COMPANIES['GOOGL'], color='red')
plt.xlabel(LST_COMP_CORR[1])
plt.ylabel('GOOGL')
plt.title('Prices')
plt.grid()
plt.show()
```

**Based on this diagram, no correlation can be claimed between HPQ and GOOGL stock prices. Their price movements are likely driven by different factors and may be considered independent.**

```python
plt.scatter(x=DF_COMPANIES[LST_COMP_CORR[2]], y=DF_COMPANIES['GOOGL'], color='0.5')
plt.xlabel(LST_COMP_CORR[2])
plt.ylabel('GOOGL')
plt.title('Prices')
plt.grid()
plt.show()
```

**Based on this diagram, no correlation can be claimed between HPQ and GOOGL stock prices. Their price movements are likely driven by different factors and may be considered independent.**

## Stage 6

```python
S_MEANS_COMP = DF_COMPANIES.mean(axis=1)
```

```python
S_MEANS_COMP
```

**Based on this data we can track monthly increases or decreases in average prices of stocks.**

## Stage 7

```python
plt.figure(figsize=(20, 7))

plt.plot(DF_COMPANIES.index, DF_COMPANIES['GOOGL'], '-bo', label='GOOGL')
plt.plot(DF_COMPANIES.index, DF_COMPANIES[LST_COMP_CORR[0]], '-go', label=LST_COMP_CORR[0] + ' - max')
plt.plot(DF_COMPANIES.index, DF_COMPANIES[LST_COMP_CORR[1]], '-ro', label=LST_COMP_CORR[1] + ' - min')
plt.plot(DF_COMPANIES.index, DF_COMPANIES[LST_COMP_CORR[2]], '-ro', label=LST_COMP_CORR[2] + ' - none')
plt.plot(DF_COMPANIES.index, S_MEANS_COMP, '-o', color='0.7', label='Average')

plt.xticks(rotation=60)
plt.legend(fontsize=15)
plt.title('Prices for every month of the period', fontsize=20)
plt.xlabel('Dates', fontsize=15)
plt.ylabel('Prices', fontsize=15)
plt.grid()
plt.tight_layout()
plt.show()
```

## The graph shows that:

 - The average stock price remained nearly unchanged during this period but rose slightly at the end.
 - GOOGLE's stock price closely tracked the average price of all stocks.
 - Stocks with maximum/minimum correlation to GOOGLE had the highest/lowest prices, respectively.
 - The average price was closer to the minimum, indicating that high-priced stocks were in the minority.
